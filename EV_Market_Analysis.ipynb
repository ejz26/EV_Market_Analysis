```python
#Delete Redundent and Unecessary Columns
df = df.drop(columns=['battery_capacity_kwh','charging_speed_kw','drive_type','top_speed_mph'])
```

```python
#Add More Meaningful Columns
df["range_per_kwh"] = df["range_miles"] / df["battery_capacity_kwh"]
df["price_per_mile"] = df["price_usd"] / df["range_miles"]
df["estimated_revenue"] = df["price_usd"] * df["annual_sales_units"]
```
